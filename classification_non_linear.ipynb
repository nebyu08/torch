```python
import torch
from torch.nn import functional as f
import torch.nn as nn 
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# lets set up the 
torch.manual_seed(42)
np.random.seed(42)

#number of samples
n_samples=1000

x,y=make_circles(n_samples,
                noise=0.03,
                random_state=42)
```

```python
print(f"the first value of {x[:10]}")
print(x.shape)
```

```python
#lets plot it
plt.scatter(x[:,0],x[:,1],c=y,cmap=plt.cm.RdBu)
```

```python
# # turn numbers into tensors
# train_size=int(0.8*len(x))
# x_train,y_train=x[:train_size:],y[:train_size]
# x_test,y_test=x[train_size:],y[train_size:]

# #lets display the data
# print(x_train.shape,y_train.shape)
# print(x_test.shape,y_test.shape)

# we can also use train_test splti from skleanr
x=torch.from_numpy(x).type(torch.float32)
y=torch.from_numpy(y).type(torch.float32)

x_train,x_test,y_train,y_test=train_test_split(x,y,
                                               test_size=0.2,
                                               random_state=42,
                                              )
```

```python
x_train
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
#lets assign the device values 
device="cuda" if  torch.cuda.is_available() else "cpu"
device
```

```python
x_train,y_train=x_train.to(device),y_train.to(device)
x_test,y_test=x_test.to(device),y_test.to(device)
```

# lets build the non linear model 

```python
def accuracy(y_pred,y_true):
    correct=torch.eq(y_pred,y_true).sum().item()
    acc=correct/len(y_pred)*100
    return acc
```

```python
class classification_v3(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer_1=nn.Linear(in_features=2,out_features=20)
        self.layer_2=nn.Linear(in_features=20,out_features=20)
        self.layer_3=nn.Linear(in_features=20,out_features=1)
        self.relu=nn.ReLU()
    def forward(self,x:torch.Tensor)->torch.Tensor:
        return self.layer_3(self.relu(self.layer_2(self.relu(self.layer_1(x)))))
```

```python
# lets initiate our model
model_3=classification_v3()
model_3
```

```python
tot_param=0
for param in model_3.parameters():
    tot_param+=param.numel()
print(f"the total number of parameter is {tot_param} ")
```

```python
# lets setup the optimizer and loss 
loss_fn=nn.BCEWithLogitsLoss()
optimizer=torch.optim.SGD(model_3.parameters(),lr=0.1)
```

```python
#let set the seed
torch.manual_seed(42)
torch.cuda.manual_seed(42)

# lets train the model
epochs=1000
for epoch in range(epochs):
    model_3.train()
    
    y_pred_logits=model_3(x_train).squeeze()
    loss_train=loss_fn(y_pred_logits,y_train)
    
    y_pred_train=torch.round(torch.sigmoid(y_pred_logits))
    acc_train=accuracy(y_train,y_pred_train)
    
    #lets clear the parameter
    optimizer.zero_grad()
    
    #backpropagation
    loss_train.backward()

    #update the parameter
    optimizer.step()

    #validation model
    model_3.eval()
    with torch.inference_mode():
    
        y_val_logits=model_3(x_test).squeeze()
        y_val_pred=torch.round(torch.sigmoid(y_val_logits))
        
        loss_val=loss_fn(y_val_logits,y_test)
        val_acc=accuracy(y_test,y_val_pred)
    
    if epoch%100==0:
        print(f"epoch is {epoch} | loss {loss_train:.3f},| test lossis: {loss_val:.3f} ,|training loss is {loss_train:.2f}% |  ,test acc {val_acc:.2f}% ")
```

# relu

```python
def relu(x:torch.Tensor)->torch.Tensor:
    return torch.maximum(torch.tensor(0),x)
```

## sigmoid

```python
def sigmoid(x:torch.Tensor)->torch.Tensor:
    return 1/(1+torch.exp(-x))
```
