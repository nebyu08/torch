```python
import torch
from torch import nn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

# lets make a synthetic data

```python
from sklearn.datasets import make_blobs

n_samples=1000
num_classes=4

num_features=2
x,y=make_blobs(n_samples=n_samples,centers=num_classes,
               cluster_std=1.5,
               n_features=2,random_state=42)

#turn our data into tensors
x=torch.from_numpy(x).type(torch.float32)
y=torch.from_numpy(y).type(torch.LongTensor)

#lets split our data into training and testing
x_train,x_test,y_train,y_test=train_test_split(x,y,
                                              random_state=42)
```

```python
# lets visualize our data
plt.figure(figsize=(10,7))
plt.scatter(x[:,0],x[:,1],c=y,cmap=plt.cm.RdYlBu);
```

```python
#lets make an accuracy function
def accuracy(y_pred,y_true):
    correct=torch.eq(y_pred,y_true).sum().item()
    acc=correct/len(y_pred)*100
    return acc
```

# lets build our model

```python
#but first lets move our data into device
device="cuda" if torch.cuda.is_available() else "cpu"
```

```python
class categorical_classification(nn.Module):
    def __init__(self,input_feature,out_feature,n_neurons=8):
        """
        args:
            
            input_feature:acceps number of input feature
            
            output_feature:number of classese to which to be classified into,
        
            number of neurons per layer
        
        returns:
            return an object model
        
        """
        super().__init__()
        self.linear_stack_layer=nn.Sequential(
            nn.Linear(in_features=input_feature,out_features=n_neurons),
            nn.ReLU(),
            nn.Linear(in_features=n_neurons,out_features=n_neurons),
            nn.ReLU(),
            nn.Linear(in_features=n_neurons,out_features=out_feature)
        )
    def forward(self,x:torch.Tensor)->torch.Tensor:
        return self.linear_stack_layer(x)
```

```python
x_train.shape
```

```python
input_shape=x_train.shape[1]
out_shape=num_classes
n_neurons=28

classifier=categorical_classification(input_feature=input_shape,
                                     out_feature=out_shape,
                                      n_neurons=n_neurons
                                     ).to(device)
```

```python
classifier
```

```python
# lets setup prepare loss and optmizer
loss_fn=torch.nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(classifier.parameters(),lr=0.001)
```

```python
with torch.inference_mode():
    y_out=torch.softmax(classifier(x_test),dim=1)
print(y_out[:10])
#print(loss_fn(y_out))
```

```python
#lets convert our models prediction into labels
y_pred_labels=torch.argmax(y_out,dim=1)
```

```python
y_pred_labels
```

```python
y_test
```

## the way our model works
* turn logits(raw model output) --> probability distributions(using softmax)  (y_preds) -> get the maximum value ditribution using argmax.

# lets make a training and validation loop 

```python
# lets train our model
epochs=100
for epoch in range(epochs):
    classifier.train()
    y_pred_logits=classifier(x_train)
    y_pred=torch.softmax(y_pred_logits,dim=1).argmax(dim=1)

    #lets calculate the loss
    loss=loss_fn(y_pred_logits,y_train)  #this is the loss function
    acc_train=accuracy(y_train,y_pred)  #this is the accuracy
    #lets clear our gradient
    optimizer.zero_grad()
    
    #lets backprop
    loss.backward()

    #lets update our gradinet
    optimizer.step()

    #evaluation time
    classifier.eval()
    with torch.inference_mode():
    
        y_val_logits=classifier(x_test)
        y_val_pred=torch.softmax(y_val_logits,dim=1).argmax(dim=1)
        
        y_val_loss=loss_fn(y_val_logits,y_test)
        acc_val=accuracy(y_test,y_val_pred)
    
    if epoch%10==0:
        print(f"epoch {epoch} | train loss {loss:.2f}% | accuracy is {acc_train:.3f}% | testing loss {y_val_loss:.2f}% | testing accuracy {acc_val:.2f}% ")
```
